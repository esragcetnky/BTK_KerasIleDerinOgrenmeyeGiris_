# 2.Bir kısım örneğin ekrana verilmesi

```
def ekrana_ver(index):
  fig,ax=plt.subplots(figsize=(14,14))
  x,y=10,10
  for i in range(index):
    plt.subplot(y,x,i+1)
    plt.imshow(x_train[i])
  plt.show()
```

```
ekrana_ver(10)
```
